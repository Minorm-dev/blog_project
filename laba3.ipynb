CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,

# Раздел 3: ключевые фрагменты кода

* Фабрика приложения - Создание и конфигурация Flask-приложения, регистрация Blueprints и контекстного процессора (app/init.py)

````python
from flask import Flask, render_template, session
from db import get_db_connection
from app.config import Config
from app.extensions import init_serializer

def create_app():
    app = Flask(__name__, template_folder='templates', static_folder='static')
    app.config.from_object(Config)

    init_serializer(app)

    # регистрируем blueprints
    from app.blueprints import auth, main, profile, posts, admin, guestbook
    app.register_blueprint(auth.bp)
    app.register_blueprint(main.bp)
    app.register_blueprint(profile.bp)
    app.register_blueprint(posts.bp)
    app.register_blueprint(admin.bp)
    app.register_blueprint(guestbook.bp)

    # контекстный процессор для текущего пользователя и статистики
    @app.context_processor
    def inject_user():
        user = None
        stats = {}
        conn = get_db_connection()
        if 'user_id' in session:
            user = conn.execute("SELECT * FROM users WHERE id = ?", (session['user_id'],)).fetchone()
        stats = {
            "users": conn.execute("SELECT COUNT(*) FROM users").fetchone()[0],
            "posts": conn.execute("SELECT COUNT(*) FROM posts").fetchone()[0],
            "comments": 0,
            "visits": conn.execute("SELECT visits FROM site_stats WHERE id = 1").fetchone()[0]
        }
        conn.close()
        return dict(current_user=user, stats=stats)

    # обработчик ошибки 403
    @app.errorhandler(403)
    def forbidden(e):
        return render_template('403.html'), 403

    return app
````

* Декоратор для проверки ролей (app/utils.py)

````python
from functools import wraps
from flask import session, flash, redirect, url_for, abort

def role_required(*roles):
    def decorator(f):
        @wraps(f)
        def wrapped(*args, **kwargs):
            if 'user_id' not in session:
                flash('Сначала войдите', 'warning')
                return redirect(url_for('auth.login'))
            if session.get('role') not in roles:
                abort(403)
            return f(*args, **kwargs)
        return wrapped
    return decorator
````

* Пример Blueprint - управления постами (app/blueprints/posts.py)

````python
from flask import Blueprint, render_template, request, redirect, url_for, flash, session, abort
from db import get_db_connection
from app.utils import role_required, check_author

bp = Blueprint('posts', __name__, url_prefix='/posts')

@bp.route('/create', methods=['GET', 'POST'])
@role_required('author', 'moderator', 'admin')
def create():
    """Создание нового поста."""
    if request.method == 'POST':
        title = request.form['title'].strip()
        content = request.form['content'].strip()
        published = 1 if request.form.get('published') else 0
        if not title or not content:
            flash('Заполните все поля', 'danger')
            return redirect(url_for('posts.create'))
        conn = get_db_connection()
        conn.execute(
            "INSERT INTO posts (title, content, author_id, published) VALUES (?, ?, ?, ?)",
            (title, content, session['user_id'], published)
        )
        conn.commit()
        conn.close()
        flash('Пост создан!', 'success')
        return redirect(url_for('main.index'))
    return render_template('create_post.html')
````

* Функция работы с базой данных (db.py)

````python
import sqlite3

DATABASE = 'database.db'

def get_db_connection():
    """Создаёт соединение с SQLite и возвращает объект connection с row_factory = sqlite3.Row."""
    conn = sqlite3.connect(DATABASE)
    conn.row_factory = sqlite3.Row
    return conn

def init_db():
    """Инициализирует таблицы и начальные данные."""
    conn = get_db_connection()
    conn.executescript("""
        CREATE TABLE IF NOT EXISTS users (...);
        CREATE TABLE IF NOT EXISTS posts (...);
        CREATE TABLE IF NOT EXISTS site_stats (...);
        CREATE TABLE IF NOT EXISTS guestbook (...);
        INSERT OR IGNORE INTO site_stats (id, visits) VALUES (1, 0);
    """)
    conn.close()
````

* Аутентификация - регистрация и логин (app/blueprints/auth.py)

````python
from werkzeug.security import generate_password_hash, check_password_hash
from db import get_db_connection

@bp.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        username = request.form['username'].strip()
        email = request.form['email'].strip()
        password = request.form['password'].strip()
        password_hash = generate_password_hash(password)
        conn = get_db_connection()
        try:
            conn.execute(
                "INSERT INTO users (username, email, password_hash) VALUES (?, ?, ?)",
                (username, email, password_hash)
            )
            conn.commit()
            flash('Регистрация прошла успешно!', 'success')
            return redirect(url_for('auth.login'))
        except Exception:
            flash('Пользователь уже существует!', 'danger')
        finally:
            conn.close()
    return render_template('register.html')
````

## Проблемы и их решения в приложении

| Проблема | Решение |
|:---------|:--------|
| **Хранение паролей в открытом виде** — угроза безопасности при утечке БД | Использование `werkzeug.security.generate_password_hash()` и `check_password_hash()` для хеширования паролей (PBKDF2). Пароли не хранятся в исходном виде. |
| **Несанкционированный доступ к маршрутам** (например, создание/удаление постов любым пользователем) | Внедрён декоратор `@role_required('author', 'moderator', 'admin')`, который проверяет роль пользователя в сессии и возвращает 403 при недостатке прав. |
| **Монолитная структура кода** — сложность поддержки и масштабирования | Рефакторинг на **Blueprints** (auth, main, posts, profile, admin, guestbook) и фабрику приложения `create_app()`. Все части приложения разделены по модулям. |
| **Отсутствие механизма восстановления пароля** — пользователь не может сменить забытый пароль | Реализован безопасный сброс через `URLSafeTimedSerializer` (itsdangerous). Генерация токена с привязкой к email и временем жизни (1 час), установка нового пароля. |
| **Отсутствие инициализации базы данных при первом запуске** — ошибки из-за отсутствия таблиц | Создана функция `init_db()` в `db.py`, которая проверяет и создаёт все таблицы, а также вставляет начальные данные (site_stats). Вызывается при старте приложения. |
| **Прямые ссылки на маршруты в шаблонах** — негибкость при изменении URL | Замена всех `url_for('index')`, `url_for('edit')` и т.д. на явные указания blueprints: `url_for('main.index')`, `url_for('posts.edit', id=...)`, `url_for('auth.login')` и т.д. |

* Проектирование реляционной БД и схемы SQLite (внешние ключи, рекурсивные связи, агрегатные функции)
* Реализация декораторов
* Настройка окружения

Для улучшения можно сделать следующее:
* Заменить встроенный механизм сессий на Flask-Login для более гибкого управления аутентификацией, защиты от подделки сессий и поддержки "запомнить меня".
* Внедрить CSRF-защиту с помощью Flask-WTF для всех форм (создание поста, гостевая книга, профиль), чтобы предотвратить межсайтовые подделки запросов.
* Добавить кэширование (Flask-Caching) для главной страницы и статистики — это снизит нагрузку на БД при частых посещениях (особенно при большом количестве постов).
* Реализовать полнотекстовый поиск с помощью SQLite FTS5 вместо LIKE, чтобы ускорить поиск по постам и обеспечить ранжирование результатов.

Оценка сложности работы: средняя, так как основная сложность работы заключалась в грамотной обработке запросов для авторизации и регистрации
